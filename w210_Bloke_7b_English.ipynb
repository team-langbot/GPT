```python
from langchain.llms import CTransformers
from langchain import PromptTemplate, LLMChain
from langchain.callbacks.streaming_stdout import StreamingStdOutCallbackHandler
```

```python
#from ctransformers import AutoModelForCausalLM, AutoTokenizer
```

```python
llm = CTransformers(model="TheBloke/Llama-2-7B-Chat-GGML", model_file = 'llama-2-7b-chat.ggmlv3.q2_K.bin', callbacks=[StreamingStdOutCallbackHandler()])
#llm = AutoModelForCausalLM.from_pretrained('marianbasti/Llama-2-13b-fp16-alpaca-spanish')
#llm = AutoModelForCausalLM.from_pretrained("marella/gpt-2-ggml")
```

```python
template = """
[INST] <<SYS>>
You are a helpful, respectful and honest assistant. Your answers are always brief.
<</SYS>>
{text}[/INST]
"""

prompt = PromptTemplate(template=template, input_variables=["text"])
```

```python
llm_chain = LLMChain(prompt=prompt, llm=llm)
```

```python
response = llm_chain.run('"If I had not told this story to my friends I have spent all evening with Juan." \
                        has a grammatical error, where "have" shold be "would have". \
                        What are the evidence words for this grammatical error? \
                        and what type of grammatical error this is? Please respond \
                        in "evidencec words =" and "grammatical error ="" format.')
```
